```python
import tensorflow as tf
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from PIL import Image
import pytesseract
```

```python
# Load the image using PIL
image_path = '/content/download (3).jpeg'
image = Image.open(image_path)

# Resize the image if needed
# image = image.resize((new_width, new_height))

# Convert the image to grayscale
gray_image = image.convert('L')

# Convert PIL image to NumPy array
numpy_image = tf.keras.preprocessing.image.img_to_array(gray_image)
```

```python
# Define the CNN model
model = tf.keras.Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=numpy_image.shape),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(64, activation='relu'),
    Dense(2, activation='softmax')  # Assuming binary classification (text vs. non-text)
])

# Load pre-trained weights if available
# model.load_weights('cnn_model_weights.h5')
```

```python
# Convert the NumPy array back to a PIL image
reconstructed_image = tf.keras.preprocessing.image.array_to_img(numpy_image)

# Perform OCR using Tesseract on the reconstructed image
extracted_text = pytesseract.image_to_string(reconstructed_image)
print("Extracted Text:")
print(extracted_text)
```
